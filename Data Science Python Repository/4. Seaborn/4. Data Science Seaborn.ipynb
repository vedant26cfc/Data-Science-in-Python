```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
sb.get_dataset_names()
```

```python
tips = sb.load_dataset('tips')
```

```python
tips.head(8)
```

Distplots

```python
#1 Distplots create a disbribution for univariate data
# THis means they will take in only a single column name
sb.displot(tips['total_bill'], kde=True)
```

```python
sb.distplot(tips['total_bill'],kde=False)
```

Joint Plots

```python
#Joint Plots allow you to match up two different  distplots
sb.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

Pairplot

```python
#IT plots all numerical columns of a data set pairwise
sb.pairplot(tips)
```

```python
#If in your data set you have a categorical column you can use hue to efficiently visualise the data
sb.pairplot(tips,hue='sex')
```

# Categorical Plots

## Barplots

```python
#A barplot plots categorical columns to a numerical column and gives off a result based on an aggregate function
sb.barplot(x='sex',y='total_bill',data=tips,estimator=sum)
```

```python
#By default the estimator is mean()
sb.barplot(x='sex',y='total_bill',data=tips)
```

```python
sb.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

## Countplot

This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value

```python
sb.countplot(x='sex',data=tips)
```

## Boxplot

Shows the distribution of categorical data as it facilitates comparisons between variables

```python
sb.boxplot(x="day", y="total_bill", data=tips)
```

```python
sb.boxplot(x="day", y="total_bill", hue="smoker",data=tips)
```

# MATRIX PLOTS
### Plots data as colored matrix which can be used to indicate clusters within the data, however data must be in matrix form

```python
flights = sb.load_dataset('flights')
```

```python
flights.head(10)
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
fp = flights.pivot_table(values='passengers',index='month',columns='year')
sb.heatmap(fp)
```

```python
sb.heatmap(fp, cmap='coolwarm', linecolor = 'white', linewidth = 1)
```

```python
sb.clustermap(fp, cmap = "coolwarm")
```
